```python
from github import Github
import os

# Lee el token desde una variable de entorno (recomendado)
token = os.getenv("GITHUB_TOKEN", "TU_GITHUB_TOKEN")
g = Github(token)
```

```python
print(token)
```

```python
user = g.get_user("octocat")
for repo in user.get_repos():
    print(f"{repo.full_name} — ⭐ {repo.stargazers_count}")
```

```python
import base64

repo = g.get_repo("tiangolo/fastapi")
readme = repo.get_readme()           # busca automáticamente README en la rama por defecto
content = base64.b64decode(readme.content).decode()
print(content[:5000])  # muestra los primeros 500 caracteres
```

```python
repo = g.get_repo("numpy/numpy")
commits = repo.get_commits(sha="main")  # o la rama que quieras
for commit in commits[:5]:
    print(f"{commit.sha[:7]} — {commit.commit.message.splitlines()[0]}")
```

```python
import os, requests

token = os.getenv("GITHUB_TOKEN", "TU_GITHUB_TOKEN")
headers = {"Authorization": f"token {token}"}

# Ejemplo: Listar repos de un usuario
resp = requests.get("https://api.github.com/users/octocat/repos", headers=headers)
for repo in resp.json():
    print(f"{repo['full_name']} — ⭐ {repo['stargazers_count']}")
```

```python
resp = requests.get(
    "https://api.github.com/search/code",
    headers=headers,
    params={"q": "def fetch+language:python", "sort": "indexed", "order": "desc"}
)
print(resp.json())
for item in resp.json().get("items", [])[:5]:
    print(f"{item['repository']['full_name']} — {item['name']} ({item['path']})")
```
